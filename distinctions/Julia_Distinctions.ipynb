## Getting Started

### The Julia REPL

Stands for the **R**ead-**E**valuate-**P**rint **L**oop

```
$ julia
               _
   _       _ _(_)_     |  Documentation: https://docs.julialang.org
  (_)     | (_) (_)    |
   _ _   _| |_  __ _   |  Type "?" for help, "]?" for Pkg help.
  | | | | | | |/ _` |  |
  | | |_| | | | (_| |  |  Version 1.12.6 (2026-04-09)
 _/ |\__'_|_|_|\__'_|  |  Official https://julialang.org release
|__/                   |

julia> 

```

Exit the REPL with `exit()` or <kbd>Ctrl</kbd>-<kbd>D</kbd>.

### REPL Modes

For each special mode, you can return to the normal Julia prompt with <kbd>backspace</kbd>.

#### Help Mode

Use the <kbd>?</kbd> key to enable, then enter a search term.

````julia
?∂
````

````julia
# Others to try:

# ?sin
# ?varinfo
# ?+
# ?@main
````

Return to Julia prompt with <kbd>backspace</kbd>.

#### Shell Mode

Use the <kbd>;</kbd> key to enable, then use (subset of) shell commands.

````julia
;pwd
````

````julia
;ls
````

Return to Julia prompt with <kbd>backspace</kbd>.

#### Pkg Mode

Use the <kbd>]</kbd> key to enable; command-style version of `Pkg` functions for package management. We'll see this more later.

````julia
]status
````

## An Expeditious Grammar Sample

Manual: [Control Flow](https://docs.julialang.org/en/v1/manual/control-flow/),
[Scopes](https://docs.julialang.org/en/v1/manual/variables-and-scoping/)

- Consistent use of `end` keyword (rather than braces or tabs)
- Unicode allowed in code (use \ and \<tab\>)
- No "switch" syntax

````julia
@show typeof(ARGS) # global container for command-line arguments

function foo(Ω::Int64)
    flag = false
    for i in 1:Ω
        if flag
            println("Breaking at i = ", i)
            break
        elseif i < Ω/2
            continue
        else
            flag = true
        end
    end
    return nothing
end

try
    foo(10)
    # error("this is an error")
catch e
    println(e.msg)
    rethrow()
end

; # this is just for the notebook
````
